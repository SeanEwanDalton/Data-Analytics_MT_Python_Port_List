Oh no! our vessel MT Python's auxiliary engine has failed! The vessel is currently with power in the English Channel. Authorities have said we must divert asap to any port that could conduct emergency repairs. 


After receiving a [list of ports](https://www.kaggle.com/datasets/sanjeetsinghnaik/ship-ports) worldwide from our management, we will need to narrow down the possibilities. 

The port list was loaded into Google Big Query in order to drill down the options via SQL. The criteria was as follows:

1.   Provide a list of 10 ports nearby the English Channel 
2.   Arrange via port size so we can assess which port might have space
